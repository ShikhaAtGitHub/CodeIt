```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def delete_head(self):
        temp = self.head
        self.head = self.head.next
        temp.next = None

        del temp

    def delete_tail(self):
        # check if list is empty
        if self.head is None:
            print("list is empty")
            return
        
        # check if there is only one node
        if self.head.next is None:
            temp = self.head
            self.head = None
            del temp
            return
        
        # but what if there is more than a single node
        # traverse the list to find the second last element
        temp = self.head
        while temp.next and temp.next.next: # temp.next.next -> since we are trying to find the second last element
            temp = temp.next

        tail = temp.next
        temp.next = None
        del tail

    def delete_kth(self, k):
        if self.head is None:
            print("List is empty")
            return
        
        # what if k is 0
        if k == 0:
            temp = self.head
            self.head = self.head.next
            temp.next = None
            del temp
            return
        
        count = 0
        prev = self.head
        while prev.next and prev.next.next and count < k - 1:
            prev = prev.next
            count += 1

        temp = prev.next
        prev.next = temp.next
        temp.next = None
        del temp

    def delete_by_val(self, val):
        if self.head is None:
            print("List is empty")
            return        
        
        # if head itself contains a value
        if self.head.val == val:
            temp = self.head
            self.head = self.head.next
            temp.next = None
            del temp
            return
        
        # traverse the list
        prev = self.head
        curr = self.head.next
        while curr:
            if curr.val == val:
                prev.next = curr.next
                curr.next = None
                del curr
                return
            prev = curr
            curr = curr.next

    def insert_at_position(self, k, val):
        new_node = Node(val)

        # Insert at head (k == 0)
        if k == 0:
            new_node.next = self.head
            self.head = new_node
            return
        
        temp = self.head
        count = 0
        while temp is not None and count < k - 1:
            temp = temp.next
            count += 1

        new_node.next = temp.next
        temp.next = new_node

    def reverse_ll(self):
        prev = None
        current = self.head

        while current:
            next_node = current.next   # Store next node
            current.next = prev        # Reverse current node's pointer
            prev = current             # Move prev to current
            current = next_node        # Move to next node

        self.head = prev

    def detect_loop(self):
        visited = set()
        temp = self.head

        while temp:
            if temp in visited:
                return True
            visited.add(temp)
            temp = temp.next
        return False
    
    # another way
    def detect_loop(self):
        slow = self.head
        fast = self.head

        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                return True
            
    # To find starting point of loop in Linked List:
```
