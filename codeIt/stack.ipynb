```python
# Stack using Array

class Stack:
    def __init__(self, capacity):
        self.stack = [None] * capacity
        self.top = -1
        self.capacity = capacity

    def is_empty(self):
        return self.top == -1
    
    def is_full(self):
        return self.top == self.capacity - 1
    
    def push(self, value):
        if self.is_full():
            print("Stack Overflow! Cannot push:", value)
            return
        
        self.top += 1
        self.stack[self.top] = value

    def pop(self):
        if self.is_empty():
            print("Stack Underflow! Cannot pop.")
            return None
        popped_value = self.stack[self.top]
        self.top -= 1
        return popped_value
    
    def peek(self): # Return the top element without removing it
        if self.is_empty():
            print("Stack is empty.")
        else:
            return self.stack[self.top]
        
    def size(self):
        return self.top + 1
    
    def display(self): # in a way that is going to be read as a stack
        if self.is_empty():
            print("Stack is empty.")
        else:
            return self.stack[:self.top + 1][::-1]
```

```python
# Balanced Parenthesis

s1 = '()[{}()]', s2 = '()[{}(])'
s3 = '(()', s3 = '))'

def is_balanced(string):
    hashmap = {}
    hashmap[')'] = '('
    hashmap['}'] = '{'
    hashmap[']'] = '['

    stack = []

    for s in string:
        if s in hashmap.values():
            stack.append(s)
        else:
            if not stack or stack[-1] != hashmap[s]:
                return False
            stack.pop()

    return len(stack) == 0
```

```python
# Next Greater Element:

arr = [4, 5, 2, 10, 8]

def next_greater_element(arr):
    output = [-1] * len(arr)
    stack = []

    for i in range(len(arr) - 1, -1, -1): # -> O(n) this one is for arr

        while stack and arr[i] > stack[-1]: # -> O(n) this one iterates for stack
            stack.pop()
        
        if stack:
            output[i] = stack[-1] # top element of stack becomes the next greater element of stack

        stack.append(arr[i])

    return output

# TC -> O(2n)
# SC -> O(2n): stack + output
```

```python
# Remove K didgits:

def remove_k_digits(num, k):
    stack = []

    for digit in num:
        while stack and k > 0 and stack[-1] > digit:
            stack.pop()
            k -= 1
        stack.append(digit)

    while stack and k > 0:
        stack.pop()
        k -= 1

    result = "".join(stack).lstrip('0')

    return result if result else '0'
```
