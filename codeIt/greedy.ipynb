```python
class Solution:
    def removeOverlap(self, intervals) -> int:
        # sort the intervals by their end time
        intervals.sort(key = lambda x: x[1])

        count = 0
        prev_end = float('-inf')

        for i in intervals:
            start, end = i
            if start >= prev_end:
                prev_end = end
            else:
                count += 1

        return count
```

```python
# Minimum no. of platforms required
# Here we are not concerned about the intervals of (arrival_time, departure_time)

def noOfPlatforms(arr, dep, n):
    arr.sort()
    dep.sort()

    i = 1
    j = 0
    curr_platform = 1
    max_platform = 1

    while i < n and j < n:
        if arr[i] <= dep[j]:
            curr_platform += 1
            i += 1
        else:
            curr_platform -= 1
            j += 1
        max_platform = max(max_platform, curr_platform)

    return max_platform
```

```python
# Fractional Knapsack

class Item:
    def __init__(self, value, weight):
        self.value = value
        self.weight = weight

class Solution:
    def fractionalKnapsack(self, W, arr):
        items = [Item(val, wt) for val, wt in arr]
        items.sort(key = lambda x: x.value/x.weight, reverse=True)
        curr_weight = 0
        final_value = 0.0
        for i in range(len(items)):
            if curr_weight + items[i].weight <= W:
                curr_weight += items[i].weight
                final_value += items[i].value
            else:
                remain = W - curr_weight
                to_be_added = (items[i].value / items[i].weight) * remain
                final_value += to_be_added
                break
        return final_value
```

```python
# Coin Changer

class CoinChanger:
    def __init__(self, coins):
        self.coins = sorted(coins, reverse = True)

    def get_min_coins(self, amount):
        coin_count = 0
        # result = []

        for coin in self.coins:
            while amount >= coin:
                num = amount // coin
                amount -= num * coin
                coin_count += num
                # result.extend([coin] * num)

        if amount != 0:
            return -1
        return coin_count
```

```python
intervals = [[1,3], [6,9]]
newInterval = [2,5]

def insert_intervals(intervals, newInterval):
    res = []
    inserted = False
    
    for curr in intervals:
        # detect left part - curr ends before newInterval starts
        if curr[1] < newInterval[0]:
            res.append(curr)

        # detect right most part - curr interval starts after newInterval ends
        elif curr[0] > newInterval[1]:
            res.append(newInterval)
            inserted = True
            res.append(curr)
        else:
            # overlapping intervals detected: minimum of starting point of curr and newInterval
            newInterval[0] = min(curr[0], newInterval[0])
            newInterval[1] = max(curr[1], newInterval[1])

    if not inserted:
        res.append(newInterval)

    return res
        
```
