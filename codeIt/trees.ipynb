```python
# longest Chain in a Binary Tree

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root):
        self.root = root
        self.max_chain = 0

    def longest_chain(self, node):
        if node is None:
            return 0
        
        left = 0
        current = node.left
        while current:
            left += 1
            current = current.left

        right = 0
        current = node.right
        while current:
            right += 1
            current = current.right

        chain_len = left + right + 1
        self.max_chain = max(self.max_chain, chain_len)

        self.longest_chain(node.left)
        self.longest_chain(node.right)

        return self.max_chain
```

```python
# Diameter of a Binary Tree

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root):
        self.root = Node(root)
        self.max_diameter = 0

    def diameter(self, root):
        def dfs(curr_node):
            if not curr_node:
                return 0

            left = self.diameter(curr_node.left)
            right = self.diameter(curr_node.right)

            self.max_diameter = max(self.max_diameter, left + right)

            return 1 + max(left, right)
        dfs(root)
        return self.max_diameter 
```

```python
# Check if Balanced Binary Tree

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def is_balanced(root):
    def check(node):
        if node is None:
            return 0

        left_depth = check(node.left)
        right_depth = check(node.right)

        if left_depth == -1 or right_depth == -1: return -1
        if abs(left_depth - right_depth) > 1: return -1

        return 1+ max(left_depth, right_depth)
    return check(root) != -1
```

```python
# Zig-Zag

from collections import deque

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def zigzag_level_order(root):
    if not root:
        return []
    
    result = []
    queue = deque([root])
    left_to_right = 0

    while queue:
        level = []

        for i in range(len(queue)):
            node = queue.popleft()
            level.append(node.val)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        if not left_to_right:
            level.reverse()

        result.append(level)
        left_to_right = not left_to_right

    return result
```

```python
# Boundary Traversal

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class Solution:
    def isLeaf(self, node):
        return not node.left and not node.right
    
    def left_boundary(self, node, res):
        curr = node.left
        while curr:
            if not self.isLeaf(curr):
                res.append(curr.val)
            if curr.left:
                curr = curr.left
            else:
                curr = curr.right 

    def right_boundary(self, node, res):
        curr = node.right
        temp = []
        while curr:
            if not self.isLeaf(curr):
                temp.append(curr.val)

            if curr.right:
                curr = curr.right
            else:
                curr = curr.left
        res.extend(reversed(temp))

    def leaves(self, node, res):
        if self.isLeaf(node):
            res.append(node.data)
            return
        
        if node.left:
            self.leaves(node.left, res)
        if node.right:
            self.leaves(node.right, res)

    def print_res(self, root):
        res = []
        if not root:
            return res
        
        if not self.isLeaf(root):
            res.append(root.data)

        self.left_boundary(root, res)
        self.leaves(root, res)
        self.right_boundary(root, res)
```

```python
# Binary tree from preorder and inorder traversals
from typing import List

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class Solution:
    def buildTree(self, preorder: List[int], inorder:List[int]):
        inorder_map = {value: idx for idx, value in enumerate(inorder)}

        def helper(pre_start, pre_end, in_start, in_end):

            root_val = preorder[pre_start]
            root = Node(root_val)

            in_root_idx = inorder_map[root_val]
            left_subtree_len = in_root_idx - in_start

            # build tree from preorder
            root.left = helper(pre_start + 1, pre_start + left_subtree_len, in_start, in_root_idx - 1)
            # pre_start + left_subtree_len + 1 -> right child begining
            root.right = helper(pre_start + left_subtree_len + 1, pre_end, in_root_idx + 1, in_end)

            return root
        return helper(0, len(preorder) - 1, 0, len(inorder) - 1)
```

```python
# Construct BST from pre-order
# Brute Force takes time complexity of O(n^2), another approach: pre-order can be sorted into in-order (TC: O(NlogN) + O(N))
# Optimal Approach: Range [min-max]
class Node:
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None

class Solution:
    def bstFromPreorder(self, preorder):
        self.index = 0
        return self._construct(preorder, float('-inf'), float('inf'))
    
    def _construct(self, preorder, lower, upper):
        if self.index == len(preorder):
            return None

        val = preorder[self.index]
        if val < lower or val > upper: # re-check this condition
            return None

        # Construct current node
        root = Node(val)
        self.index += 1

        # Construct left and right subtrees using bounds
        root.left = self._construct(preorder, lower, val)
        root.right = self._construct(preorder, val, upper)

        return root
    
    def construct(self, preorder, lower, upper):
        val = preorder[self.index]
        if val > lower and val < upper:
            return None
        root = Node(val)
        self.index += 1

        root.left = self.construct(preorder, lower, val)
        root.right = self.construct(preorder, val, upper) # since I am moving to right so I need to make sure that the node is greater than the minimal value.
        return root
```

```python
# Morris Inorder Traversal

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def morris_inorder_traversal(root):
    curr = root
    result = []

    while curr:
        # Case 1: No left child
        if curr.left is None:
            result.append(curr.val)
            curr = curr.right
        else:
            # Find inorder predecessor
            pre = curr.left
            while pre.right and pre.right != curr:
                pre = pre.right

            ## Case 2a: Thread doesn't exist create it
            if pre.right is None:
                pre.right = curr
                curr = curr.left
            # Case 2b: Thread exists -> remove it
            else:
                pre.right = None
                result.append(curr.val)
                curr = curr.right

    return result
```

```python
# Iterative Preorder Traversal: Root -> Left -> Right

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root):
        self.root = Node(root)

    def iterative_preorder(self):
        if not self.root:
            return []
        
        stack = [self.root]
        result = []

        while stack:
            curr = stack.pop()
            result.append(curr.val)

            if curr.right:
                stack.append(curr.right)
            if curr.left:
                stack.append(curr.left)

        return result
```

```python
# Height of a Binary Tree using BFS

from collections import deque

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root):
        self.root = Node(root)

    def height(self):
        if not self.root:
            return 0
        
        queue = deque([self.root])
        height = 0

        while queue:
            height += 1
            for _ in range(len(queue)):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
        
        return height
```

```python
# Check for a Balanced Binary Tree

# if a tree is balanced binary tree then I will return height of the tree, else I will return -1

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root):
        self.root = Node(root)

    def is_balanced(self):
        if self.root is None:
            return 0

        left_height = self.height(self.root.left)
        right_height = self.height(self.root.right)

        if left_height == -1 or right_height == -1:
            return -1
        if abs(left_height - right_height) > 1:
            return -1
        return max(left_height, right_height) + 1
```

```python
# Diameter of a Binary Tree

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self, root):
        self.root = Node(root)
        self.diameter = float('-inf')

    def height(self, node):
        if not node:
            return 0
        
        left_height = self.height(node.left)
        right_height = self.height(node.right)

        self.diameter = max(self.diameter, left_height + right_height) # update the maximum diameter found so far
        return max(left_height, right_height) + 1
    
    def get_diameter(self):
        self.height(self.root)
        return self.diameter # globally calculated diameter
```
